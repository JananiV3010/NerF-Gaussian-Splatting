```python
import os

DATA_ROOT = "data/books_processed_debug2"
ABSOLUTE_DATA_ROOT = os.path.abspath(DATA_ROOT)
DB_PARENT_PATH = os.path.join(ABSOLUTE_DATA_ROOT, "colmap")
DB_PATH = os.path.join(DB_PARENT_PATH, "database.db")
IMG_PATH = os.path.join(ABSOLUTE_DATA_ROOT, "images")

print(f"Absolute data root: {ABSOLUTE_DATA_ROOT}")
print(f"DB Parent Path: {DB_PARENT_PATH}")
print(f"DB Path: {DB_PATH}")
print(f"Image Path: {IMG_PATH}")

# Ensure the colmap directory is clean and exists
!rm -rf "$DB_PARENT_PATH"
!mkdir -p "$DB_PARENT_PATH"

# Verify directory creation and permissions
!ls -ld "$DB_PARENT_PATH"
!touch "$DB_PARENT_PATH/test.txt"
!rm "$DB_PARENT_PATH/test.txt"

# Run colmap feature_extractor in headless mode, disabling GPU for SIFT
%env QT_QPA_PLATFORM=offscreen
!colmap feature_extractor \
  --database_path "$DB_PATH" \
  --image_path "$IMG_PATH" \
  --ImageReader.single_camera 1 \
  --ImageReader.camera_model OPENCV \
  --SiftExtraction.use_gpu 0 \
  --SiftExtraction.estimate_affine_shape 0 \
  --SiftExtraction.domain_size_pooling 0
```

```python
!colmap exhaustive_matcher \
  --database_path "$DB_PATH" \
  --SiftMatching.use_gpu 0
```

```python
SPARSE_DIR="$DATA_ROOT/colmap/sparse"
!mkdir -p "$SPARSE_DIR"

!colmap mapper \
  --database_path "$DB_PATH" \
  --image_path "$IMG_PATH" \
  --output_path "$SPARSE_DIR"
```
